# Part 2: Data Cleaning and Feature Engineering

Next, we will create a new DataFrame that uses a combination of features from the existing data and engineered features that provide standard baseball statistics.

```python
df = pd.DataFrame(data[['Salary', 'Years', 'AtBat', 'CAtBat', 'Runs', 'CRuns', 'RBI', 'CRBI']])

df['Average'] = np.round(data['Hits'] / data['AtBat'], 3)
df['CAverage'] = np.round(data['CHits'] / data['CAtBat'], 3)

df['OBP'] = np.round(data[['Hits', 'Walks']].sum(axis=1) / data[['AtBat', 'Walks']].sum(axis=1), 3)
# OBP is not perfect, no data on HBP or SF
df['COBP'] = np.round(data[['CHits', 'CWalks']].sum(axis=1) / data[['CAtBat', 'CWalks']].sum(axis=1), 3)

# Percentage of home runs per at bat
df['HR/AB'] = np.round(data['HmRun'] / data['AtBat'], 5)
df['CHR/AB'] = np.round(data['CHmRun'] / data['CAtBat'], 5)

# Number of at bats per year, higher number likely indicates a larger salary
df['AB/Yr'] = np.round(df['CAtBat'] / df['Years'], 0).astype(int)


df['FieldingAttempts'] = data[['PutOuts', 'Assists', 'Errors']].sum(axis=1)
df['FieldingPct'] = np.round((data['PutOuts'] + data['Assists']) / df['FieldingAttempts'], 3)

# Need to consider DH, replace fielding percentage with 0, model should learn that this number
# is arbitrary if FieldingAttempts is 0
df['FieldingPct'] = np.where(df['FieldingAttempts'] == 0, 0., df['FieldingPct'])

# Integer encode the catagorical features
# If-else works here as there are only two categories for each feature
df['League'] = data['League'].apply(lambda x: 1 if x == 'N' else 0) # 0 = AL, 1 = NL
df['Division'] = data['Division'].apply(lambda x: 1 if x == 'E' else 0) # 0 = West, 1 = East
df['NewLeague'] = data['NewLeague'].apply(lambda x: 1 if x == 'N' else 0) # 0 = AL, 1 = NL

df.head(20)
```

```python
correlations = df.corr()
plt.figure(figsize=(12, 10))
plt.title('Correlation Matrix of Features')
sns.heatmap(correlations, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.show()
```

At bats per year shows to have a high correlation with the salary at 0.5. All of the cumulative statistics are higher than the other column which also makes sense since it is a better representation of them as a player as opposed to a single season. The hitting statistics are much more correlated with the salary than the fielding statistics.

# Part 3: Modeling

We will use a few different models to predict the salary of the players. We score the models buy the root mean squared error (RMSE).

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Create a root mean squared error function
def rmse(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

# Split up the data
X, Y = df.drop(['Salary'], axis=1), df['Salary']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

predictions = pd.DataFrame(y_test)
```

## Linear Regression

We will start with the simplest and most interpretable model, Linear Regression.

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)

lr_pred = lr.predict(x_test)
predictions['LinearRegression'] = lr_pred

print(f'       Intercept: {lr.intercept_}', end='\n\n')

for coef, col in zip(lr.coef_, x_train.columns):
    print(f'{col:>16}: {coef}')

print('\nLinear Regression RMSE:', rmse(y_test, lr_pred))
```

The Linear Regression RMSE is not great at coming around 368.

## Random Forest Regression

We will now use the ensemble method Random Forest Regression.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

rf_parameters = {'n_estimators': [100, 200, 300, 500], 'max_depth': [2, 4, 6, 8, 10], 'max_features': [2, 4, 6, 8, 10, None]}

# Use 5-fold cross validation to find the best parameters scored on mean squared error
rf_grid = GridSearchCV(RandomForestRegressor(), rf_parameters, cv=5, scoring='neg_mean_squared_error')
rf_grid.fit(x_train, y_train)
rf_best_params = rf_grid.best_params_
rf_best_params
```

Using these parameters we can build our random forest model using the best hyperparameters found in the grid search.

```python
rf = RandomForestRegressor(**rf_best_params, random_state=0)
rf.fit(x_train, y_train)
rf_pred = rf.predict(x_test)
predictions['RandomForest'] = rf_pred
print('Random Forest RMSE:', rmse(y_test, rf_pred))
```

The Random Forest RMSE is doing better than the Linear Regression RMSE. Now we will look at SHAP values to determine feature importance.

```python
import shap
explainer = shap.TreeExplainer(rf)
shap_values = explainer.shap_values(x_test)

plt.figure(figsize=(12, 10))
plt.title('SHAP Values for Random Forest')
shap.summary_plot(shap_values, x_test)
plt.show()
```

You can see that cumulative at bats is the most important feature, which shows that the more at bats they get tends to increase their salary since this shows both experience and a lot of playing time.

At bats per year is also an important feature which shows that in general if the number is higher the player is likely to have a higher salary (with a few exceptions).

Cumulative home runs per at bats is also important, which shows that those that hit for power tend to get a higher salary as a result.

Years is also an interesting feature because the model finds that those with little experience (number of years) tend to have a salary cut and those with the highest number of years played have a salary increase; however, those with a number of years between the two generally see the highest salary. This could be because they are most likely to be playing in their prime.

Cumulative Average and cumulative on base percentage are also important.

The fielding features are not nearly as important as hitting.

The three catagorical variables are seen as the least important, with NewLeague being the most important and if NewLeague is National League then the player is most likely to have a slightly higher salary.

## XGBoost Regression

For our final model we will use XGBoost Regression, an ensemble method that is similar to Random Forest Regression but is considered to be more powerful. It starts with weak trees that iteratively learn over time. Compared to the random forest--which uses several randomly generated decision trees--XGBoost learns which features are important in each iteration and adjusts each tree in order to better predict the target.

We will set up a similar grid search to the random forest model, but use an XGBoost model instead.

```python
from xgboost import XGBRegressor
xgb_parameters = {'max_depth': [2, 4, 6, 8, 10], 'n_estimators': [100, 200, 300, 500], 'learning_rate': [0.001, 0.01, 0.05, 0.1], 'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0]}
xgb_search = GridSearchCV(XGBRegressor(), xgb_parameters, cv=5, scoring='neg_mean_squared_error')
xgb_search.fit(x_train, y_train)
xgb_best_params = xgb_search.best_params_
xgb_best_params
```

```python
xgb = XGBRegressor(**xgb_best_params, random_state=0)
xgb.fit(x_train, y_train)
xgb_pred = xgb.predict(x_test)
predictions['XGBoost'] = xgb_pred
print('XGBoost RMSE:', rmse(y_test, xgb_pred))
```

You can see the XGBoost RMSE is better than the Random Forest RMSE. We will also look at SHAP values to determine feature importance, and see if they are different from the Random Forest.

```python
xgb_explainer = shap.TreeExplainer(xgb)
xgb_shap_values = xgb_explainer.shap_values(x_test)

plt.figure(figsize=(12, 10))
plt.title('SHAP Values for XGBoost')
shap.summary_plot(xgb_shap_values, x_test)
plt.show()
```

All of the values are roughly the same (the cumulative statistics are generally the most important, the fielding statistics are not as important and the catagorical features are the least important).
